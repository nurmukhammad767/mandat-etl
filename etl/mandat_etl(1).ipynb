```python
from pyspark.sql.functions import col, when, regexp_replace
from pyspark.sql.types import FloatType
```

```python
df=spark.read.option("header","true") \
        .option("inferschemaa","true") \
        .csv("Files/mandat_uzbmb_uz_2025/*.csv")
```

```python
df=df.withColumn("TR class", when(col("TR class")=='table-success', 'Grant')
                             .when(col("TR class")=='table-warning', 'Kontrakt')
                             .otherwise("Yiqilgan")
)
```

```python
df = df.withColumn("Ball", regexp_replace(col("Ball"), ",", ".").cast("float"))
```

```python
df=df.drop("Batafsil")
```

```python
df=df.withColumnRenamed("TR class", "Status") \
     .withColumnRenamed("F.I.SH", "Full_Name")\
     .withColumnRenamed("Oliy ta'lim muassasasi", "University") \
     .withColumnRenamed("Yo'nalish", "Major") \
     .withColumnRenamed("Ta'lim tili", "Language") \
     .withColumnRenamed("Ta'lim shakli", "format") \
     .withColumnRenamed("Ball", "Score")

     	
```

```python
# JDBC connection details
jdbc_hostname = "mandat-server.database.windows.net"
jdbc_port = 1433
database_name = "mandat_databse"
jdbc_url = f"jdbc:sqlserver://{jdbc_hostname}:{jdbc_port};databaseName={database_name}"
username = '' 
password = ''      
table_name = "abituriyent_2024"


df.write \
    .format("jdbc") \
    .option("url", jdbc_url) \
    .option("dbtable", table_name) \
    .option("user", username) \
    .option("password", password) \
    .mode("overwrite") \
    .save()

print(f"Table '{table_name}' created and DataFrame written successfully.")
```
